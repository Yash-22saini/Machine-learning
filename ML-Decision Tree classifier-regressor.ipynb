# Decision tree

```python
# a decision tree is a supervised machine learning model that is used for classification and regression class that is why it is considered as classification and regression tree
# It works like a flowchart or a tree structure where each internal node representt a feature and each branch represent a decision or condition and each leaf node represent an output
#  and class label in the classification or a continouus value in the regression
     
```

```python
# A decision tree splits the data into subsets based in the value of input features and this process is recussive forming a tree like structure

# HOW IT WORKS:---
# 1. it start with the entire data---> it choose the best feature to split the data based on a matrix like gini index, entropy, mse (mean suqared error ) then it will split the dataset into smaller subsets
# then it will repeat the process for each child node untill a stopping condition is met.
```

# Decision tree classifier

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("covid_toy.csv")
df.head(2)
```

```python
df = df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['gender']= lb.fit_transform(df['gender'])
df['cough']= lb.fit_transform(df['cough'])
df['city']= lb.fit_transform(df['city'])
df['has_covid']= lb.fit_transform(df['has_covid'])
```

```python
df.head(2)
```

```python
x = df.drop(columns = ['has_covid'])
y = df['has_covid']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,
                                                    random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train ,y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

# Decision tree regressor

```python
df = pd.read_csv("insurance.csv")
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['sex']= lb.fit_transform(df['sex'])
df['smoker']= lb.fit_transform(df['smoker'])
df['region']= lb.fit_transform(df['region'])
```

```python
x = df.drop(columns = ['charges'])
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train , y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test , y_pred)
```
