```python
# (2). LogisticRegression Model ---- > when we have target data is categorical .
# Backend ---- > Activation_function(sigmoid) ---- > probability basis
# Output (0-1) .
# threshold_value = 0.5 or 1/2
# if value > threshold_value:
#return 1
# else:
#return 0
```

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("covid_toy.csv")
df.head(2)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer()
```

```python
df['fever'] = si.fit_transform(df[['fever']])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb= LabelEncoder()
```

```python
df['gender']= lb.fit_transform(df['gender'])
df['cough']= lb.fit_transform(df['cough'])
df['city']= lb.fit_transform(df['city'])
df['has_covid']= lb.fit_transform(df['has_covid'])
```

```python
df.head(2)
```

```python
x = df.drop(columns = ['has_covid'])
y = df['has_covid']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,
                                                    random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test) 
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

# ATTRITION DATA LOGISTIC REGRESSION

```python
df = pd.read_csv("Attrition.csv")
df.head(2)
```

```python
#df.isnull().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb= LabelEncoder()
```

```python
df['Attrition']= lb.fit_transform(df['Attrition'])
df['BusinessTravel']= lb.fit_transform(df['BusinessTravel'])
df['Department']= lb.fit_transform(df['Department'])
df['EducationField']= lb.fit_transform(df['EducationField'])
df['Gender']= lb.fit_transform(df['Gender'])
df['JobRole']= lb.fit_transform(df['JobRole'])
df['MaritalStatus']= lb.fit_transform(df['MaritalStatus'])
df['Over18']= lb.fit_transform(df['Over18'])
df['OverTime']= lb.fit_transform(df['OverTime'])
```

```python
#df.dtypes
```

```python
df.head(2)
```

```python
x = df.drop(columns = ['Attrition'])
y = df['Attrition']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,
                                                    random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test) 
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```
