# Model training

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb= LabelEncoder()
```

```python
df['airline']= lb.fit_transform(df['airline'])
df['flight']= lb.fit_transform(df['flight'])
df['source_city']= lb.fit_transform(df['source_city'])
df['departure_time']= lb.fit_transform(df['departure_time'])
df['stops']= lb.fit_transform(df['stops'])
df['arrival_time']= lb.fit_transform(df['arrival_time'])
df['destination_city']= lb.fit_transform(df['destination_city'])
df['class']= lb.fit_transform(df['class'])
df['duration']= lb.fit_transform(df['duration'])
```

```python
df.head(5)
```

```python
x = df.drop(columns = ['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(
    n_estimators=100, 
    max_features="sqrt", 
    max_depth=6, 
    max_leaf_nodes=6
)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test) 
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test , y_pred)
```

```python
import joblib
```

```python
joblib.dump(rf,'rf_model.pkl')
print("model exported as rf_model.pkl sucessfully")
```
