# Ai_Assitant data is classification data  You have to apply LogisticRegression , RandomForestClassifier , Naive Bayes on AI_Assitant data.

# logistic regression

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("ai_assistant_usage_student_life.csv")
df.head(2)
```

```python
#df.isnull().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb= LabelEncoder()
```

```python
df['SessionID']= lb.fit_transform(df['SessionID'])
df['StudentLevel']= lb.fit_transform(df['StudentLevel'])
df['Discipline']= lb.fit_transform(df['Discipline'])
df['SessionDate']= lb.fit_transform(df['SessionDate'])
df['SessionLengthMin']= lb.fit_transform(df['SessionLengthMin'])
df['TaskType']= lb.fit_transform(df['TaskType'])
df['FinalOutcome']= lb.fit_transform(df['FinalOutcome'])
df['UsedAgain']= lb.fit_transform(df['UsedAgain'])
df['SatisfactionRating']= lb.fit_transform(df['SatisfactionRating'])
```

```python
df.dtypes
```

```python
df.head(2)
```

```python
x = df.drop(columns = ['UsedAgain'])
y = df['UsedAgain']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,
                                                    random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test) 
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

# Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test) 
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

# Naive bayes

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
x_train_1 = sc.fit_transform(x_train)
```

```python
x_test_1 = sc.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
nb.fit(x_train_1,y_train)
```

```python
y_pred = nb.predict(x_test_1) 
```

```python
y_pred
```
