```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y= make_regression(n_samples=4, n_features = 1, n_informative=1, n_targets= 1 , noise = 80, random_state=13)
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(x, y)
plt.plot(x, lr.predict(x), color = 'red')
```

```python
# Now we are seperating the line
```

```python
y_pred=((78.35*x)+0).reshape(4)
```

```python
plt.scatter(x, y)
plt.plot(x, lr.predict(x), color='red', label="OLS")
plt.plot(x,y_pred, color='green', label="b=0")
plt.legend()
plt.show()
```

```python
m=78.35
b=0
loss_slope = -2*np.sum(y-m*x.ravel()-b) ####dl/db
loss_slope
```

```python
#Learning rate=0.1
lra=0.1
step_size=loss_slope*lra
step_size
```

```python
# new intercept
b=b-step_size   # b_mew =b_old-lr*dl/db
b
```

```python
y_pred1=((78.35*x)+b).reshape(4)

plt.scatter(x, y)
plt.plot(x, lr.predict(x), color = 'black', label='OLS')
plt.plot(x, y_pred, color='red', label =' b={}'.format(b))
plt.plot(x, y_pred1, color="yellow", label = 'b=0')
plt.legend()
plt.show()
```

```python
# Iteration -2
loss_slope=-2*np.sum(y -m*x.ravel()-b)
loss_slope
```

```python
step_size = loss_slope * lra
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((78.35*x)+b).reshape(4)

plt.scatter(x, y)
plt.plot(x, lr.predict(x), color = 'black', label='OLS')
plt.plot(x, y_pred, color='red', label =' b={}'.format(b))
plt.plot(x, y_pred1, color="yellow", label = 'b=20')
plt.plot(x, y_pred2, color="blue", label = 'b=25')
plt.legend()
plt.show()
```

```python
# Iteration -3
loss_slope=-2*np.sum(y -m*x.ravel()-b)
loss_slope
```

```python
step_size = loss_slope * lra
step_size
```

```python
y_pred3=((78.35*x)+b).reshape(4)

plt.scatter(x, y)
plt.plot(x, lr.predict(x), color = 'black', label='OLS')
plt.plot(x, y_pred, color='red', label =' b={}'.format(b))
plt.plot(x, y_pred1, color="yellow", label = 'b=0')
plt.plot(x, y_pred2, color="blue", label = 'b=20')
plt.plot(x, y_pred3, color="green", label = 'b=25')

plt.legend()
plt.show()
```
