```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/spotify-youtube-data/Spotify Youtube Dataset.csv")
df.head(2)
                
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df[df.select_dtypes('float64').columns] = df.select_dtypes('float64').fillna(df.select_dtypes('float64').mean())
df[df.select_dtypes('object').columns] = df.select_dtypes('object').fillna(df.select_dtypes('object').mode().iloc[0])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = df.select_dtypes(include='object').columns

lb = LabelEncoder()

for col in categorical_cols:
    df[col] = lb.fit_transform(df[col])
```

```python
df.head(5)
```

```python
x = df.drop(columns = ['Stream'])
y = df['Stream']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr= LinearRegression()

lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test) 
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test) 
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test , y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train , y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test , y_pred)
```
